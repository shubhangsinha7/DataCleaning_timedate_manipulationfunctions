```python
first_qua = float(df.salary.quantile([0.25])) #finding the first quartile
third_qua = float(df.salary.quantile([0.75])) #finding the third quartile 
#any value outside of these two quartiles would be defined as an outlier 
df_new = df[(df.salary<=first_qua)]
df_new = df[(df.salary>=third_qua)]     
df_new.describe()
 #standard deviation of salary is now aprox. half of the intial value hence the outliers have been removed 
```
